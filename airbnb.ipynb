```python
import pandas as pd
df = pd.read_excel("1730285881-Airbnb_Open_Data.xlsx")
df.head()
```

```python
# Check shape (rows, columns)
print("Shape:", df.shape)

# Check column names
print("Columns:", df.columns)

# Data types & missing values
print(df.info())

# Quick stats
print(df.describe(include="all"))
```

```python
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]', '', regex=True).astype(float)

print("Average price:", df['price'].mean())
print("Average service fee:", df['service fee'].mean())
```

```python
# Remove $ and commas, convert to float
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]', '', regex=True).astype(float)

# Convert dates
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Convert price and service fee to numeric
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]', '', regex=True).astype(float)

# Convert review date
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')

# Fill missing values (optional)
df['reviews per month'] = df['reviews per month'].fillna(0)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='room type', data=df, palette="Set2")
plt.title("Room Type Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='room type', y='price', data=df, palette="coolwarm")
plt.title("Price Distribution by Room Type")
plt.ylim(0, 1000)  # optional: limit extreme prices
plt.show()
```

```python
plt.figure(figsize=(6,6))
df['neighbourhood group'].value_counts().plot(kind="pie", autopct='%1.1f%%')
plt.title("Neighbourhood Group Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['price'], bins=50, kde=True, color="purple")
plt.title("Price Distribution")
plt.xlabel("Price ($)")
plt.show()
```

```python
reviews_per_year = df.groupby(df['last review'].dt.year)['id'].count()

plt.figure(figsize=(8,5))
reviews_per_year.plot(kind="line", marker='o')
plt.title("Number of Reviews per Year")
plt.xlabel("Year")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='availability 365', y='price', data=df, alpha=0.5)
plt.title("Availability vs Price")
plt.xlabel("Availability (days per year)")
plt.ylabel("Price ($)")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
